Regresion lineal
==

- Realize una regresion lineal sobre los datos de penales pateados (_PKatt_) y penales anotados (_PK_).

```python
from sklearn import linear_model
import numpy as np
import sklearn.metrics as sm

X=[]
for i in messi_csv.PKatt:
    X.append(i)

Y=[]
for j in messi_csv.PK:
    Y.append(j)

x=np.array(X)
y=np.array(Y)

print(x)
print(y)
print("--------")

#Modifico la estructura del array

x_mod=x.reshape(len(x),1)
y_mod=y.reshape(len(y),1)

#Regresion Lineal

Regresor=linear_model.LinearRegression()
Regresor.fit(x_mod,y_mod)
Y_predicha=Regresor.predict(x_mod) #Aca realiza la regresion con los datos ingresados al objeto Regresor

print("Error absoluto medio =", round(sm.mean_absolute_error(y_mod, Y_predicha), 2)) 
print("Error cuadratico medio =", round(sm.mean_squared_error(y_mod, Y_predicha), 2)) 
print("Error absoluto mediano =", round(sm.median_absolute_error(y_mod, Y_predicha), 2)) 
print("Puntuación de varianza explicada =", round(sm.explained_variance_score(y_mod, Y_predicha), 2)) 
print("Puntuación R2 =", round(sm.r2_score(y_mod, Y_predicha), 2))



plot1=plt.figure(figsize=(8,8))
plt.scatter(messi_csv['PKatt'], messi_csv['PK'])
plt.xlabel('Penales pateados p/ competencia')
plt.ylabel('Penales anotados p/ competencia')
plt.plot(x_mod, Y_predicha , color='red', linewidth=1)

plt.show()
```

- Realize una regresion lineal **contraída** sobre los datos de minutos jugados (_Minutes played_) y goles anotados (_Goals scored_). ¿Que tan bien ajusta el modelo a los datos de entrenamiento? Utilize metricas de sklearn para evaluar este ajuste.
- Si Messi fuese a jugar 1500 minutos en alguna competicion, ¿cuantos goles anotaría?. Realize una prediccion con el modelo entrenado.

```python
from sklearn import linear_model


#Renombro el header del archivo csv para eliminar los espacios en blanco (da error si no al llamarla)

messi_csv.rename(columns={"Minutes played":"Minutes_played","Goals scored":"Goals_scored"},inplace=True)

X=[]
Y=[]

for i in messi_csv.Minutes_played:
    X.append(i)

Y=[]
for j in messi_csv.Goals_scored:
    Y.append(j)

x=np.array(X)
y=np.array(Y)

#Divido los datos en entrenamiento y prueba

datos_entrenamiento = int(0.8 * len(X))
print(datos_entrenamiento)
datos_prueba = len(X) - datos_entrenamiento
print(datos_prueba)

#Entrenamiento

x_entrenamiento = x[:datos_entrenamiento].reshape((datos_entrenamiento,1))
y_entrenamiento = y[:datos_entrenamiento].reshape((datos_entrenamiento,1))

#Prueba

x_prueba = x[datos_entrenamiento:].reshape((datos_prueba,1))
y_prueba = y[datos_entrenamiento:].reshape((datos_prueba,1))


#Se realiza la regresion

regresion_ridge = linear_model.Ridge(alpha=0.01, fit_intercept=True, max_iter=10000)
regresion_ridge.fit(x_entrenamiento,y_entrenamiento)  #le damos los datos al objeto regresion_ridge de tipo clase

y_predicha_ridge=regresion_ridge.predict(x_prueba)

print("Error absoluto medio =", round(sm.mean_absolute_error(y_prueba, y_predicha_ridge), 2)) 
print("Error cuadratico medio =", round(sm.mean_squared_error(y_prueba, y_predicha_ridge), 2)) 
print("Error absoluto mediano =", round(sm.median_absolute_error(y_prueba, y_predicha_ridge), 2)) 
print("Puntuación de varianza explicada =", round(sm.explained_variance_score(y_prueba, y_predicha_ridge), 2)) 
print("Puntuación R2 =", round(sm.r2_score(y_prueba, y_predicha_ridge), 2))

x_1500m=np.array(1500).reshape(-1,1)
y_predicho_1500m=int(regresion_ridge.predict(x_1500m))
print("Goles anotados predichos en 1500 min = ",y_predicho_1500m)

plt.figure(figsize=(8,8))
plt.scatter(messi_csv['Minutes_played'],messi_csv['Goals_scored'])
plt.xlabel('Minutos jugados p/ competencia')
plt.ylabel('Goles anotados p/ competencia')
plt.plot(x_prueba, y_predicha_ridge , color='red', linewidth=1)
plt.show()
```

Clasificacion
==

- Utilizando regresion logistica clasifique los datos de asistencias (_Assists_) y goles (_Goals scored_) en **2 clusters diferentes.** Para esto, debera separar los datos en 2 categorerias (por ejemplo, "0" y "1"). Se sugiere clasificar los puntos con asistencias mayores a a 7,5 en un cluster y los restantes en otro cluster.

```python
messi_csv.rename(columns={"Goals scored":"Goals_scored"},inplace=True)

X=[]
Y=[]

for i in messi_csv.Assists:
    X.append(i)

Y=[]
for j in messi_csv.Goals_scored:
    Y.append(j)

Z=[]

for k in (messi_csv.Assists):
    if k<7.5:
        Z.append(0)
    elif k>=7.5:
        Z.append(1)

X_Y=[]
for i in range(len(X)):
    X_Y.append([X[i],Y[i]])


x=np.array(X)
y=np.array(Y)
x_y=np.array(X_Y)
z=np.array(Z)

print(x)
print(y)
print(z)
print(x_y)

#Se realiza la regresion logistica

clasificador = linear_model.LogisticRegression(solver='lbfgs', C=100)
clasificador.fit(x_y, z)
prediccion = clasificador.predict(x_y)
print(prediccion)
print("Puntaje de clustering=",clasificador.score(x_y,z)) 

#Graficamos

x_min,x_max=min(x_y[:,0])-1,max(x_y[:,0])+1
y_min,y_max=min(x_y[:,1])-1,max(x_y[:,1])+1

step_size=0.01

valor_x,valor_y=np.meshgrid(np.arange(x_min,x_max,step_size),np.arange(y_min,y_max,step_size)) #arange crea una secuencia de num con un salto especifico
print("///")
print(valor_x)
print(valor_y)
malla_puntos=clasificador.predict(np.c_[valor_x.ravel(),valor_y.ravel()])  #np._c funciona como un reshape, donde ordena las listas dadas en dos columnas
print("valor_x\n",valor_x)
print("valor_y\n",valor_y)
print("Malla sin cambio de forma")
print(malla_puntos)
malla_puntos=malla_puntos.reshape(valor_x.shape) #cambiamos la forma de la malla a la misma que los valores x e y
print("Malla con forma cambiada")
print(malla_puntos)
plt.figure()
plt.pcolormesh(valor_x,valor_y,malla_puntos,cmap=plt.cm.gray,shading="auto")
plt.scatter(x_y[:,0],x_y[:,1],c=clasificador.predict(x_y),s=80,edgecolors="black",linewidths=1,cmap=plt.cm.Paired) #le doy los valores con el modelo ya entrenado para que le de los colores
#el modelo se entreno con un conjunto de datos pero le estoy dando los mismos datos para que prediga, deberia haber separado los datos en entrenamiento y prueba

plt.figure(figsize=(8,8))
plt.scatter(messi_csv['Assists'],messi_csv['Goals_scored'])
plt.xlabel('Asistencias de goles p/ competencia')
plt.ylabel('Goles anotados p/ competencia')
plt.show()
```

- Utilizando K-NN clasifique los datos de partidos jugados (_Matches Played_) y minutos jugados (_Minutes played_) en **4 clusters diferentes.** 
- Grafique las areas pertenencientes a cada categoria.
- Grafique con distintos colores las labels: 
    * Copa del Rey
    * Supercopa de España
    * La Liga
    * Champions League

```python
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

messi_csv.rename(columns={"Matches Played":"Matches_Played"},inplace=True)

n_neighbors = 15
X=[]
Y=[]
Z=[]
X_Y=[]

for i in messi_csv.Matches_Played:
    X.append(i)

for j in messi_csv.Minutes_played:
    Y.append(j)

for k in range(len(X)):
    X_Y.append([X[k],Y[k]])

for l in messi_csv.Competition:
    if l=="Champions League":
        Z.append(0)
    elif l=="La Liga":
        Z.append(1)
    elif l=="Copa del Rey":
        Z.append(2)
    else:  #no lee la ñ de supercopa de españa, puedo probar cambiando el encoding o esto
        Z.append(3)


x_y=np.array(X_Y)
z=np.array(Z)
print("X_y\n",x_y)
print("z\n",z)
#Creo el mapa de colores y el paso para el mallado.

h = 1 #Eligo un paso alto porque si es muy chico el proceso de arange deja sin memoria y se crashea.
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF',"#FAFFAA"])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF',"#F0FF00"])

print(cmap_light.N)
#Hago un bucle for para darle distintos pesos a los puntos, uno que tengan peso uniforme y otro variable

for weights in ["uniform","distance"]:
    clf=neighbors.KNeighborsClassifier(n_neighbors,weights=weights)
    clf.fit(x_y,z)

    #Creo los limites del grafico
    x_min,x_max=min(x_y[:,0])-1,max(x_y[:,0])+1
    y_min,y_max=min(x_y[:,1])-1,max(x_y[:,1])+1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    L= clf.predict(np.c_[xx.ravel(), yy.ravel()])

    #Agregamos al grafico
    L=L.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx,yy,L,shading="auto",cmap=cmap_light,vmin=0,vmax=4) #colormaps se normalizan entre 0 y 1,0 seria el primer color del mapa y 1 el ultimo.Con vmin y vmax se tiene en cuenta esto (ver)
    puntos=plt.scatter(x_y[:,0],x_y[:,1],c=z,cmap=cmap_bold,edgecolors="black",s=20,linewidths=1)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.legend(handles=puntos.legend_elements()[0],labels=["Champion League","La Liga","Copa del rey","Supercopa de España"])
    plt.title("Clasificación (k = %i, weights = '%s')" % (n_neighbors, weights))
plt.show()


plt.figure(figsize=(8,8))
plt.scatter(messi_csv['Matches_Played'],messi_csv['Minutes_played'])
plt.xlabel('Partidos jugados p/ competencia')
plt.ylabel('Minutos jugados p/ competencia')
plt.show()
```

- Utilizando arboles de decision clasifique los datos de partidos jugados (_Matches Played_) y goles anotados (_Goals scored_) en **4 clusters diferentes.** 
- Grafique las areas pertenencientes a cada categoria.
- Grafique con distintos colores las labels:

    * La Liga

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import graphviz


x_entrenamiento,x_test,y_entrenamiento,y_test=train_test_split(messi_csv[["Matches_Played","Minutes_played"]],z)
arbol=DecisionTreeClassifier(max_depth=3)
arbol.fit(x_entrenamiento, y_entrenamiento)
arbol.score(x_test, y_test)
arbol.score(x_entrenamiento, y_entrenamiento)
export_graphviz(arbol, out_file='arbol1.dot', class_names=["Champion League","La Liga","Copa del rey","Supercopa de España"], feature_names=["Matches_Played","Minutes_played"], impurity=False, filled=True)
with open('arbol1.dot') as f:
    dot_graph=f.read()
graphviz.Source(dot_graph)
```
